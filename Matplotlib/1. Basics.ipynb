# **Функция plt.plot(). Линия графика, точки на линии графика, несколько линий на одном графике**

## **Линия графика**

Общепринятая команда для импорта matplotlib: `import matplotlib.pyplot as plt`.

Если передать в качестве аргумента обычный список Python, то его значения будут интерпретированы как значения `y`, а индексы - как значения `x` (пример). Можно передавать в качестве аргументов серии Pandas (пример ниже).

**Дополнительные аргументы:**
* `linewidth`: толщина линии.
* `linestyle`: тип линии. Может принимать следующие значения: `'-'`, `'--'`, `'-.'`, `':'`.
* `color`: цвет линии.
* `alpha`: яркость линии. Принимает значения от 0 до 1.

Параметр `color` принимает rgb-коды и названия разнообразных цветов на английском языке, например: `'white'`, `'black'`, `'yellow'`, `'orange'`, `'red'`, `'green'`, `'blue'` и др.  Также можно использовать названия тёмных или светлых оттенков цветов, например: `'darkgreen'`, `'darkred'`, `'lightyellow'`, `'lightgrey'`. Есть еще возможность использовать различные оттенки стандартных цветов, с такими названиями как: `'snow'`, `'silver'`, `'gold'`, `'olive'`, `'lime'`, `'coral'`, `'royalblue'` и др.

Несколько стандартных цветов можно использовать в сокращённом варианте: `'k'` - чёрный, `'w'` - белый, `'r'` - красный, `'y'` - жёлтый, `'g'` - зелёный, `'b'` - синий, `'c'` - цвет морской волны, `'m'` - пурпурный.

**Примечание:** вместо параметров: `color`, `linewidth` и `linestyle` можно использовать их сокращенные наименования: `c`, `lw` и `ls`, соответственно.

```python
# линейный график на основе обычного списка Python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
# линейный график с кастомизированной линией
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5], c='royalblue', ls='-.', lw=2)
plt.show()
```

Функция `plt.show()` принимает на вход и значения датафрейма Pandas. Первым аргументом передаются данные, которые должны отложиться на графике по оси абцисс, вторым - данные, которые должны отложиться по оси ординат.

```python
import matplotlib.pyplot as plt
import pandas as pd

values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

# print(df)

#     x   y
# 0   2   0
# 1   4   5
# 2   8   4
# 3  12  11
# 4  15  12
# 5  20  10
# 6  25   9

plt.plot(df['x'], df['y'])
plt.show()
```

```python
# ещё пример графика без дополнительного форматирования
import matplotlib.pyplot as plt
import pandas as pd

values = [['Австралия', 25], ['Азия', 4540], ['Африка', 1280],
          ['Европа', 742], ['Северная Америка', 543], ['Южная Америка', 428]]
columns = ['Часть света', 'Численность в 2018 г., млн чел.']
df = pd.DataFrame(values, columns=columns)

plt.plot(df['Часть света'], df['Численность в 2018 г., млн чел.'])
plt.xticks(rotation=45)
plt.show()
```

## **Точки на линии графика**
На графике можно выделять точки при помощи параметра `marker`.

Значения параметра `marker`:
* `'o'` - точки в виде кругов.
* `'.'` - точки в виде маленьких точек.
* `'*'` - точки в виде звёздочек.
* `'s'` - точки в виде квадратов.
* `'x'` - точки в виде тонкого наклонённого крестика.
* `'X'` - точки в виде жирного наклонённого крестика.
* `'P'` - точки в виде жирного крестика.
* `'p'` - точки в виде пятиугольника.
* `'D'` - точки в виде ромба.
* `'<'`, `'>'`, `'^'`, `'v'` - точки в виде наконечника стрелок в соответствующем направлении.

Для изменения размера точки используется параметр `markersize`.

С помощью параметра `markeredgecolor` можно делать контур точек другого цвета. Параметр `markeredgewidth` регулирует толщину внешнего контура.

С помощью параметра `markerfacecolor` можно делать заполнение контура другим цветом.

**Примечание:** на графике можно оставить только точки, а линию не отображать вообще (параметру `linestyle` присвоить значение `None`). Тогда получится диаграмма рассеяния.

```python
# пример графика со стандартными точками
import matplotlib.pyplot as plt
import pandas as pd

values = [['Австралия', 25], ['Азия', 4540], ['Африка', 1280],
          ['Европа', 742], ['Северная Америка', 543], ['Южная Америка', 428]]
columns = ['Часть света', 'Численность в 2018 г., млн чел.']
df = pd.DataFrame(values, columns=columns)

plt.plot(df['Часть света'], df['Численность в 2018 г., млн чел.'], marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
# пример графика со кастомизированными точками
import matplotlib.pyplot as plt
import pandas as pd

values = [['Австралия', 25], ['Азия', 4540], ['Африка', 1280],
          ['Европа', 742], ['Северная Америка', 543], ['Южная Америка', 428]]
columns = ['Часть света', 'Численность в 2018 г., млн чел.']
df = pd.DataFrame(values, columns=columns)

plt.plot(df['Часть света'], df['Численность в 2018 г., млн чел.'], marker='s', markeredgecolor='black', markerfacecolor='red')
plt.xticks(rotation=45)
plt.show()
```

## **Несколько линий на графике**

На графике можно отобразить несколько линий, если передавать функции `plt.plot()` попарно значения для оси абцисс и оси ординат. Либо использовать несколько функций `plt.plot()`, каждая из которых будет отвечать за построение одной линии графика (`plt.show()` нужно выводить после задания всех линий). Второй вариант имеет смысл, если необходимо кастомизировать дизайн линий/точек.

```python
# пример графика с несколькими линиями, задаваемыми через одну функцию plt.plot()
import matplotlib.pyplot as plt
import pandas as pd

values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)

plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], marker='o')
plt.show()
```

```python
# пример графика с несколькими кастомизированными линиями, задаваемыми через несколько функций plt.plot()
import matplotlib.pyplot as plt
import pandas as pd

values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)

plt.plot(df['x1'], df['y1'], marker='o', color='darkblue')
plt.plot(df['x2'], df['y2'], marker='d', color='black')
plt.plot(df['x1'], df['y1'] + 2, marker='P', color='red')
plt.show()
```

# **Функция plt.title(). Заголовок графика**

В библиотеке matplotlib за отображение заголовка графика отвечает функция `plt.title()`. Принимает на вход строку с названием графика.

# **Функции plt.xlabel(), plt.ylabel(). Названия осей**

Осям графика присваивают наименования с помощью функций `plt.xlabel()` и `plt.ylabel()`. Функциям в качестве аргумента передаётся строка с именем соответствующей оси.

**Дополнительные аргументы:**
* `fontsize`: размер шрифта.
* `color`: цвет шрифта.
* `rotation`: наклон (по оси X по умолчанию - 0, по оси Y - 90).
* `labelpad`: отвечает за степень близости расположения подписей к осям.

```python
# дефолтные подписи осей графика
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5])
plt.grid(color='purple', linestyle='dotted', linewidth=1, axis='x')
plt.title('График зависимости X от Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# кастомизированные подписи осей график
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5])
plt.grid(color='purple', linestyle='dotted', linewidth=1, axis='x')
plt.title('График зависимости X от Y', color='purple', fontsize=14)
plt.xlabel('X', fontsize=14, color='purple', labelpad=20)
plt.ylabel('Y', fontsize=14, color='purple', labelpad=20, rotation=0)
plt.show()
```

# **Функция plt.legend(). Легенда на графике**

Функция `plt.legend()` позволяет отображать легенду на графике. Для того, чтобы её вывести, можно либо передать в функцию `plt.plot()` параметр `label` с указанием названий для меток легенды, а затем вывести её при помощи функции `plt.legend()`, либо передать названия для меток легенды напрямую в функцию `plt.legend()` в виде списка.

По умолчанию Matplotlib сам выбирает оптимальное место на графике для отображения легенды. Но есть возможность принудительно его изменить при помощи параметра `loc` функции `plt.legend()`. Параметр может принимать значения:
* `'best'` или `0` (по умолчанию).
* `'upper right'`  или `1`.
* `'upper left'`  или `2`.
* `'lower left'` или `3`.
* `'lower right'` или `4`.
* `'right'` или `5`.
* `'center left'` или `6`.
* `'center right'` или `7`.
* `'lower center'` или `8`.
* `'upper center'` или `9`.
* `'center'` или `10`.

Параметру `loc` также можно передавать список или кортеж с относительными координатами, куда следует поместить легенду. Относительными координаты являются потому, что размеры осей абсцисс и ординат в данном случае принимают значения в одну единицу. Первое значение в списке указывает на местоположение легенды относительно оси абсцисс, а второе - расположение легенды относительно оси ординат.

**Дополнительные параметры:**
* `ncols`: количество колонок, используемых в легенде (по умолчанию - 1).
* `facecolor`: цвет заливки области легенды.
* `edgecolor`: цвет обрамления области легенды.
* `fontsize`: размер шрифта легенды.
* `labelcolor`: цвет текста легенды.

```python
# пример стандартной легенды
import matplotlib.pyplot as plt
import pandas as pd

values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)

plt.plot(df['x1'], df['y1'], marker='o', label='x1_y1')
plt.plot(df['x2'], df['y2'], marker='o', label='x2_y2')
plt.grid(ls=':', alpha=0.5)
plt.legend()
plt.show()
```

```python
# пример ручного выбора расположения легенды
import matplotlib.pyplot as plt
import pandas as pd

values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)

plt.plot(df['x1'], df['y1'], marker='o')
plt.plot(df['x2'], df['y2'], marker='o')
plt.grid(ls=':')
plt.legend(['x1_y1', 'x2_y2'], loc=(1.05, 0.45), edgecolor='black')
plt.style.use('ggplot')
plt.show()
```

# **Функция plt.subplot(). Построение сразу нескольких графиков**

В ситуациях, когда нужно построить графики сразу на нескольких координатных осях, применяется функция `plt.subplot()`. Этой функции передаются следующие параметры: общее количество рядов, общее количество столбцов и индекс, на котором будет находиться конкретный график.

Чтобы не было наложений названий и индексов, нужно перед выводом прописать функцию `plt.tight_layout()`.

Для того, чтобы задать название получившейся группе графиков, используется функция `plt.suptitle()`.

**Примечание:** такой подход к построению графиков на одном рисунке удобен тогда, когда требуется построить симметричное количество подграфиков по рядам и столбцам, например, 2х2, 3х3, 4х4 и т.д. Когда же подграфики не симметричны по рядам и столбцам, то логика индексации подграфиков становится запутанной и менее интуитивно понятной.

```python
# пример построения сразу нескольких графиков
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 5)
y1 = x**2
y2 = x**3
y3 = x**0.5
y4 = np.sin(x**2)

plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.grid(ls=':', alpha=0.5)
plt.title('y = $x^2$')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.grid(ls=':', alpha=0.5)
plt.title('y = $x^3$')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'y')
plt.grid(ls=':', alpha=0.5)
plt.title('y = $sqrt(x)$')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'c')
plt.grid(ls=':', alpha=0.5)
plt.title('y = sin($x^2$)')

plt.suptitle('Графики')
plt.style.use('default')
plt.tight_layout()
plt.show()
```

В примере выше для вывода на экран четырёх графиков четыре раза использовась функция `plt.subplot()`. В каждой из этих функций первым аргументом было указано, что на экран следует вывести графики в два ряда, вторым аргументом - что графики должны отобразиться двумя столбцами, и третим аргументом - на какой именно позиции следует вывести данный конкретный подграфик. И далее под каждой функцией `plt.subplot()` прописывалась та функциональность, которая относилась к этому подграфику (параметры линий, сетка, название).
