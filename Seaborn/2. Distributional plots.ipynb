# **Функция sns.kdeplot(). Диаграмма KDE**

**Диаграмма KDE** (**K**ernel **D**ensity **E**stimate, ядерная оценка плотности) строит сглаженную оценку плотности вероятности для непрерывных данных. Это позволяет визуально оценить, как значения распределены по всему диапазону данных. Диаграммы KDE в библиотеке seaborn строятся при помощи функции `sns.kdeplot()`.

**Параметры:**
* `data`: DataFrame, массив с данными.
* `x`: имя переменной, которая будет отображена на оси `X`.
* `y`: имя переменной, которая будет отображена на оси Y. Если задан, строится 2D KDE.
* `hue`: имя переменной для группировки данных по категориям, каждая из которых будет представлена отдельной линией KDE.
* `palette`: палитра цветов для различения категорий, заданных параметром `hue`.
* `hue_order`: список для определения порядка категорий, отображаемых при использовании `hue`.
* `hue_norm`: два значения, определяющие нормализацию данных по оси `hue`.
* `common_norm`: логическое значение, определяющее, нормализовать ли плотности по всем категориям `hue` вместе (`True`) или отдельно (`False`). По умолчанию `common_norm=True`.
* `common_grid`: логическое значение, определяющее, использовать ли общую сетку для всех категорий `hue`. По умолчанию `common_grid=False`.
* `color`: цвет линий и заливки, если `hue` не используется.
* `linewidth`: толщина линии.
* `weights`: массив значений для взвешивания точек данных. Должен быть такой же длины, как и данные.
* `fill`: логическое значение, определяющее, следует ли заполнять область под KDE.
* `multiple`: определяет как отображать несколько KDE: `'layer'`, `'stack'` или `'fill'` (по умолчанию `'layer'`).
* `cumulative`: логическое значение, определяющее, строить ли накопительную функцию распределения (по умолчанию `cumulative=False`).
* `bw_method`: метод для вычисления ширины окна (`bandwidth`): `'scott'` (по умолчанию), `'silverman'` или числовое значение.
* `bw_adjust`: множитель для ширины окна. Умножает значение, вычисленное `bw_method`, на этот коэффициент (по умолчанию `bw_adjust=1`).
* `log_scale`: Логическое значение или кортеж из двух логических значений для использования логарифмической шкалы по осям `X` и/или `Y`.
* `levels`: количество уровней контуров для 2D KDE (по умолчанию `levels=10`).
* `thresh`: минимальное значение плотности для отображения контуров в 2D KDE (по умолчанию `thresh=0.05`).
* `gridsize`: количество точек в сетке для оценки плотности (по умолчанию `gridsize=200`).
* `cut`: количество стандартных отклонений для увеличения границ осей при построении KDE (по умолчанию `cut=3`). Указывает, насколько далеко за пределы данных следует отображать кривую KDE.
* `clip`: пределы значений для обрезки KDE. Принимает кортеж из двух чисел или двух кортежей (для двумерных данных), определяющий минимальное и максимальное значение для оси или осей.
* `legend`: логическое значение, определяющее, отображать ли легенду (по умолчанию `legend=True`).
* `cbar`: логическое значение, определяющее, отображать ли цветовую шкалу (по умолчанию `cbar=False`).

```python
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='flipper_length_mm', hue='island')

plt.show()
```

```python
# пример построения дефолтного kde-графика
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(penguins['body_mass_g'])

plt.show()
```

```python
# пример построения кастомизированного kde-графика
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(penguins['body_mass_g'],
            fill=True,
            color='pink',
            lw=3)

plt.show()
```

```python
# пример построения накопительного графика оценки плотности ядра
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='body_mass_g',
            cumulative=True)

plt.show()
```

```python
# пример построения графика с узкой полосой пропускания
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='body_mass_g',
            fill=True,
            bw_method=0.1) # добавление bw_adjust=2 чуть сгладит

plt.show()
```

```python
# пример разделения данных на группы и построения для каждой своего kde-графика
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='body_mass_g',
            fill=True,
            hue='species')
plt.show()
```

```python
# пример настройки отображения графиков плотности при помощи параметра multiple
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='body_mass_g',
            fill=True,
            hue='species',
            multiple='stack')

plt.show()
```

```python
# пример построения дефолтного 2D KDE
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')

plt.show()
```

```python
# пример кастомизированного графика 2D KDE
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
            fill=True,
            cbar=True)

plt.show()
```

```python
# пример построения кастомизированного 2D KDE с разбиением на группы и установлением порогового уровня плотности
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
            hue='species',
            fill=True,
            palette='cubehelix',
            thresh=0.2)

plt.show()
```

# **Функция sns.histplot(). Гистограммы**

Для создания гистограмм в *seaborn* используется функция `sns.histplot()`.

**Параметры:**
* `data`: данные, которые мы хотим визуализировать. Обычно это датафрейм или массив.
* `x`, `y`: название столбца данных, который мы хотим использовать для построения гистограммы (если data - это DataFrame).
* `fill`: заливка графика. По умолчанию - `True`.
* `linewidth`: толщина линии (логично использовать с `fill=False`)
* `hue`: используется для добавления цвета по категории данных.
* `multiple`: определяет, как будут отображаться подгруппы, указанные с помощью параметра `hue`.
* `bins`: количество корзин (по умолчанию *seaborn* сам выбирает подходящее количество). Принимает на вход либо целое число, либо список с числами, которые определяют границы каждой корзины.
* `shrink`: позволяет уменьшить ширину бинов
* `binwidth`: определяет ширину каждого бина (корзины) в гистограмме. Каждый бин будет охватывать определённый интервал на линии `x`.
* `binrange`: этот параметр позволяет задавать диапазон значений, которые будут охвачены гистограммой. Принимает кортеж из двух чисел (min и max).
* `kde`: логический параметр, который определяет, строить ли поверх гистограммы плотность распределения (KDE - Kernel Density Estimate). По умолчанию - `False`.
* `stat`: определяет, какую статистику нужно вычислять и отображать на гистограмме.
* `cumulative`: используется для создания накопительной (кумулятивной) гистограммы (по умолчанию `cumulative=False`).
* `element`: определяет, как будет визуализироваться гистограмма. Например, `'bars'` для столбиков или `'step'` для ступенчатой линии.
* `color`: цвет гистограммы.

Чтобы построить вертикальную гистограмму, нужно передать значения параметру `y`, а не `x`.

**Возможные значения параметра `stat`:**
* `'count'`: Количество наблюдений в каждом бине (используется по умолчанию). Гистограмма показывает, сколько значений попадает в каждую корзину (бин).
* `'frequency'`: Частота наблюдений в каждом бине. Эта статистика делит количество наблюдений в каждом бине на общее количество наблюдений, показывая долю наблюдений.
* `'density'`: Указывает, что вместо количества наблюдений (то есть частоты), по оси ординат будет отображаться плотность распределения. Эта статистика нормализует гистограмму так, чтобы суммарная площадь всех баров равнялась 1. Это полезно для сравнения распределений с разным количеством данных.
* `'probability'`: Вероятность наблюдений в каждом бине. Похожа на density, но вместо плотности показывает вероятность (т.е. сумма высот всех баров равна 1).
* `'percent'`: Процент наблюдений в каждом бине. Эта статистика умножает частоту на 100, чтобы показать долю в процентах.
* `'proportion'`: Каждый бин гистограммы показывает долю данных, которые попадают в этот бин относительно общего числа данных. То есть, сумма всех высот бинов будет равна 1.

**Возможные значения параметра `element`:**
* `'bars'` (по умолчанию): Отображает бины в виде столбцов (баров). Это традиционный вид гистограммы, где высота каждого столбца соответствует количеству наблюдений или другой выбранной метрике.
* `'step'`: Отображает гистограмму в виде ступенчатого графика. В этом режиме высоты бинов соединяются прямыми линиями, создавая вид ступенек.
* `'poly'`: Отображает гистограмму в виде линейного графика. В этом режиме вершины каждого бина соединяются прямыми линиями, создавая замкнутую линию, напоминающую линейный график.

```python
# пример построения стандартной горизонтальной гистограммы
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins)

plt.show()
```

```python
# пример построения стандартной вертикальной гистограммы
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(y='bill_length_mm', data=penguins)

plt.show()
```

```python
# пример построения кастомизированной гистограммы
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins,
             color='royalblue',
             alpha=0.6,
             kde=True)

plt.show()
```

```python
# пример ручного задания количества бинов
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins,
             bins=20)

plt.show()
```

```python
# пример ручного задания границ бинов
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins,
             bins=[20, 30, 40, 50, 60, 70])

plt.show()
```

```python
# пример ручного задания ширины бина
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins,
             binwidth=5)

plt.show()
```

```python
# пример ручного задания диапазона значений, охватываемых гистограммой
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins,
             binwidth=5,
             binrange=(35, 55))

plt.show()
```

```python
# пример построения гистограммы с кумулятивным эффектом
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.histplot(x='bill_length_mm', data=penguins,
             stat='percent',
             fill=False,
             element='step',
             linewidth=4,
             cumulative=True)

plt.axhline(80, color='red') # добавим горизонтальную линию на отметке 80

plt.show()
```
