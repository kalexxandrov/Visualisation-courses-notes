**Графики отношений (relational plots)** относятся к типу графиков, которые показывают взаимосвязи между двумя или более переменными. Эти графики позволяют находить зависимости и закономерности в данных.

В библиотеке *seaborn* существует два основных типа графиков отношений:

* **Диаграмма рассеяния (scatter plot)**. Показывает взаимосвязь двух переменных в виде точек на графике. В библиотеке *seaborn* диаграммы рассеяния строятся функцией `sns.scatterplot()`.
* **Линейный график (line plot)**. Представляет собой линию, последовательно соединяющую точки данных. Он особенно эффективен для демонстрации изменений переменной во времени. В библиотеке *seaborn* линейные графики строятся функцией `sns.lineplot()`.

Также в библиотеке *seaborn* имеется универсальная функция `sns.relplot()`, которая позволяет создавать как диаграммы рассеяния, так и линейные графики.

# **Функция sns.scatterplot(). Диаграммы рассеяния**

В библиотеке seaborn за создание диаграмм рассеяния отвечает функция `sns.scatterplot()`.

**Параметры функции:**
* `data`: DataFrame или массив с данными.
* `x`: название столбца или вектор данных для оси `X`.
* `y`: название столбца или вектор данных для оси `Y`.
* `color`: цвет точек (принимает наименование цвета или его HEX-код).
* `alpha`: прозрачность точек (принимает значение от 0 до 1).
* `s`: размер точек (принимает числовое значение).
* `marker` определяет форму маркеров (точек) на графике.
* `hue`: название столбца или вектор данных для группировки точек по цвету. Может принимать как категориальную, так и количественную переменную.
* `size`: название столбца или вектор данных для изменения размера точек. Параметр наиболее полезен при работе с числовыми данными, так как размеры точек будут пропорциональны значениям этой переменной.
* `style`: название столбца или вектор данных для изменения стиля точек (например, разные формы).
* `palette`: цветовая палитра для переменной `hue`.
* `hue_order`: задаёт порядок отображения категорий для переменной `hue`. Он принимает список значений, которые определяют порядок отображения категорий, что позволяет контролировать, какие цвета будут назначены конкретным категориям.
* `hue_norm`: позволяет нормализовать данные переменной `hue`. Параметру передаётся кортеж с минимальным и максимальным значениями, в пределах которых цветовая палитра будет меняться от самого светлого тона до самого тёмного. Это может пригодиться в случаях, когда необходимо акцентировать внимание на определенном диапазоне значений.
* `sizes`: диапазон размеров для переменной `size`. Принимает на вход кортеж с указанием минимального и максимального размеров точек в случае использования параметра `size` вместе с количественной перменной или список с размерами точек - в случае использования с категориальной переменной (пример).
* `size_order`: задаёт порядок отображения категорий для переменной `size`.
* `size_norm`: позволяет нормализовать данные переменной `size`, задавая диапазон значений.
* `markers`: список стилей маркеров для переменной `style`.
* `style_order`: задаёт порядок отображения категорий для переменной `style`.

Чтобы выводить графики без текста объекта графика, нужно либо использовать функцию `plt.show()` библиотеки Matplotlib, либо ставить `;` вместо запуска функции.

Отображение координантной сетки зависит от выбранного стиля, который изменяется функцией `sns.set_style()`. Белый стиль без отображения координатной сетки называется `'white'`. Чтобы на графике отображались линии сетки координат, надо функции `sns.set_style()` передать атрибут `'whitegrid'` или `'darkgrid'`.

Некоторые из доступных кодов маркеров:
* `'o'`: круг.
* `'s'`: квадрат.
* `'^'`: треугольник вверх.
* `'v'`: треугольник вниз.
* `'>'`: треугольник вправо.
* `'<'`: треугольник влево.
* `'x'`: крестик.
* `'*'`: звёздочка.
* `'D'`: ромб.
* `'H'`: шестиугольник.

Список доступных палитр для параметра `palette` можно посмотреть при помощи команды `print(sns.palettes.SEABORN_PALETTES.keys())`. Также можно использовать цветовые карты из библиотеки *matplotlib* (их больше). Посмотреть список цветовых карт можно при помощи команды `print(plt.colormaps())`.

Для того, чтобы посмотреть, какие цвета содержатся в той или иной цветовой палитре, используется функция `sns.color_palette()`. Принимает на вход два аргумента: палитру и количество цветовых градаций. Можно вывести в виду hex-кодов при помощи команды `print(sns.color_palette('палитра', кол-во).as_hex())`.

```python
# пример дефолтной диаграммы рассеяния
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
# другой вариант передачи аргументов в функцию
# sns.scatterplot(x=penguins['bill_length_mm'], y=penguins['bill_depth_mm'])

plt.show()
```

```python
# пример кастомизированной диаграммы рассеяния
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                color='green', s=75, marker='s', alpha=0.5)

plt.show()
```

```python
# пример изменения цвета точек в зависимости от количественного признака
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                hue='body_mass_g')

plt.show()
```

```python
# пример нормализации цветовой шкалы
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

# нормализуем так, чтобы все значения больше 4000 отображались одинаковым цветом
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                hue='body_mass_g', hue_norm=(2700, 4000))

plt.show()
```

```python
# пример ручного контроля порядка категорий
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                hue='species',
                hue_order=['Gentoo', 'Chinstrap', 'Adelie'])

plt.show()
```

```python
# изменения стиля точек в зависимости от изменения категориальной переменной
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                hue='species', 
                style='species', 
                s=70,
                markers=['<', '^', '>'])

plt.show()
```

```python
# изменения размера точек в зависимости от величины количественной переменной
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins').dropna()

sns.set_style('darkgrid')

sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
                size='body_mass_g',
                sizes=(5, 150))

plt.show()
```
