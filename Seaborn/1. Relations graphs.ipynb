**Графики отношений (relational plots)** относятся к типу графиков, которые показывают взаимосвязи между двумя или более переменными. Эти графики позволяют находить зависимости и закономерности в данных.

В библиотеке *seaborn* существует два основных типа графиков отношений:

* **Диаграмма рассеяния (scatter plot)**. Показывает взаимосвязь двух переменных в виде точек на графике. В библиотеке *seaborn* диаграммы рассеяния строятся функцией `sns.scatterplot()`.
* **Линейный график (line plot)**. Представляет собой линию, последовательно соединяющую точки данных. Он особенно эффективен для демонстрации изменений переменной во времени. В библиотеке *seaborn* линейные графики строятся функцией `sns.lineplot()`.

Также в библиотеке *seaborn* имеется универсальная функция `sns.relplot()`, которая позволяет создавать как диаграммы рассеяния, так и линейные графики.

**Примечание:** так как *seaborn* построен поверх *matplotlib*, то можно комбинировать функции из обеих библиотек. К примеру, наносить сетку при помощи функции `plt.grid()` поверх графика, построенного при помощи `sns.lineplot()`.

# **Функция sns.lineplot(). Линейные графики**

За построение линейного графика в библиотеке *seaborn* отвечает функция `sns.lineplot()`.

**Параметры функции:**
* `data`: DataFrame или массив с данными.
* `x`: название столбца или вектор данных для оси `X`.
* `y`: название столбца или вектор данных для оси `Y`.
* `errorbar`: отображение диапазона стандартных ошибок. По умолчанию `errorbar=('ci', 95)`, т.е. отображается 95%-й доверительный интервал. Чтобы отключить, надо передать аргумент `None`. В более старых версиях `seaborn` параметр называется `ci` и принимает числовое значение доверительного интервала.
* `marker`: отображение точек на графике. Некоторые из доступных значений описаны в разделе про диаграммы рассеяния.
* `markersize` (`ms`): размер точек на графике.
* `markerfacecolor`: цвет точек на графике.
* `linewidth` (`lw`): толщина линии.
* `color`: цвет линии.
* `alpha`: прозрачность линии графика.
* `hue`: позволяет разделить данные на группы по заданной категориальной переменной и представить каждую группу в виде отдельной линии на графике, используя различные цвета или стили. *seaborn* автоматически обнаруживает уникальные значения этой переменной и создает отдельную линию для каждого уникального значения, используя соответствующий цвет или стиль.
* `hue_order`: позволяет задать определенный порядок для категорий переменной, которая используется в параметре `hue`. (по умолчанию *seaborn* автоматически определяет порядок категорий на основе их появления в данных). Полезно, когда есть определенный логический или предпочтительный порядок для категорий, который необходимо отразить на графике. Принимает на вход список со значениями переменной `hue` в нужном порядке.
* `palette`: используемая палитра для линий, создаваемых при разбивке на группы по заданной категориальной переменной при помощи параметра `hue`.
* `size`: позволяет указать переменную, значения которой будут определять толщину линий на графике.
* `size_order`: позволяет задать конкретный порядок для значений переменной, указанной в `size`.
* `sizes`: позволяет задавать диапазон возможных значений толщины линий, задаваемой параметром `size`. Принимает на вход либо кортеж с парой минимальное-максимальное значения, либо список с конкретными значениям для каждой из линий.

**Параметр estimator**

В случае разделения данных на группы по заданной категориальной переменной при помощи параметра `hue` seaborn автоматически производит группировку с вычислением средних значений и использует полученные значения для построения графиков. Для того, чтобы изменить функцию-агрегатор используется параметр `estimator`, который принимает встроенные функции (`'median'`, `'sum'`), функции NumPy (при условии его импорта) или пользовательские функции.

**Boostrap**

В функции `sns.lineplot()` для оценки доверительных интервалов используется метод бутстрап. **Бутстрап** (Bootstrap) - это статистический метод, используемый для оценки свойств распределения данных (например, среднего, дисперсии, доверительных интервалов) с помощью многократного повторного взятия выборок из исходных данных с возвращением (т.е. одно значение может попасть в одну подвыборку несколько раз, а какие-то могут вообще не встретиться в новой подвыборке). По умолчанию (`errorbar=('ci', 95)`) создаётся тысяча подвыборок, в которых вычисляются средние значения, после чего отсекаются 2.5% самых верхних и 2.5% самых нижних значений; оставшееся минимальное значение - нижняя граница доверительного интервала, а максимальное значение - верхняя граница. За количество создаваемых бустрап-выборок в функции `sns.lineplot()` используется параметр `n_boot` (по умолчанию равен 1000), а воспроизводимость обеспечивается параметром `seed`.

```python
# пример построения дефолтного линейного графика
import matplotlib.pyplot as plt
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

sns.lineplot(x='time', y='pulse', data=exercise)

plt.show()
```

```python
# пример построения кастомизированного графика
import matplotlib.pyplot as plt
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

sns.lineplot(x='time', y='pulse', data=exercise,
             errorbar=None,
             marker='o',
             markersize=7.5,
             markerfacecolor='black',
             linewidth=2,
             color='black')

plt.grid(linestyle='--', alpha=0.5)

plt.show()
```

```python
# пример построения отдельных линий для каждого из значений категориальной переменной
import matplotlib.pyplot as plt
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='kind',
             hue_order=['running', 'walking', 'rest'],
             errorbar=None)

plt.show()
```

```python
# пример регулирования толщины линий в зависимости от значений категориальной переменной
import matplotlib.pyplot as plt
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='kind',
             errorbar=None,
             size='kind',
             size_order=['running', 'walking', 'rest'],
             sizes=[5, 2, 2])

plt.show()
```

```python
# демонстрация логики группировки, осуществляемой seaborn по заданной категориальной переменной
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

exercise[['kind', 'time', 'pulse']].groupby(by=['kind', 'time'], observed=True).mean()
# получившиеся значения - точки для дефолтного графика
```

```python
# изменение функции-агрегатора при группировке, осуществляемой seaborn по заданной категориальной переменной
import matplotlib.pyplot as plt
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

sns.lineplot(x="time", y="pulse", data=exercise,
             hue='kind',
             errorbar=None,
             marker='o',
             estimator='median')

plt.show()
```

```python
# пример работы boostrap при расчёте доверительного интервала
import matplotlib.pyplot as plt
import seaborn as sns

exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='diet',
             marker='o',
             errorbar=('ci', 67),
             n_boot=1000,
             seed=322)

plt.show()
```

# **Функция sns.relplot(). Универсальный способ создания реляционных графиков**

Функция `sns.relplot()` библиотеки *seaborn* является универсальной функцией для визуализации отношений между переменными и используется для создания различных типов реляционных графиков, таких как линейные графики (`lineplot`) и точечные графики (`scatterplot`).

**Основные параметры функции:**
* `data`: DataFrame, массив с данными.
* `x`, `y`: имена переменных, которые будут отображены на осях X и Y, соответственно.
* `hue`: переменная для группировки данных по цвету.
* `palette`: параметр, определяющий цветовую палитру для переменной `hue`. Может быть именем палитры, списком цветов или словарем, отображающим уровни переменной в цветах.
* `hue_order`: порядок отображения уровней переменной `hue`.
* `hue_norm`: нормализация для цветовой палитры.
* `size`: переменная, значение которой будет использоваться для изменения размера точек.
* `sizes`: значения для переменной `size`, которые будут отображены в виде размеров маркеров. Может быть словарем или списком значений.
* `size_order`: порядок отображения уровней переменной `size`.
* `size_norm`: нормализация для размеров маркеров.
* `style`: переменная, значение которой будет использоваться для изменения стиля точек или линий.
* `markers`: маркеры для переменной `style`. Может быть списком или словарем.
* `dashes`: определяет типы линий (например, пунктирные) для переменной `style`. Принимает список кортежей, где каждый кортеж определяет длину штриха и пробела.
* `style_order`: определяет порядок отображения уровней переменной `style`.
* `units`: имя переменной, которая определяет семантические подгруппы для раздельного отображения каждой точки.
* `weights`: имя переменной, которая определяет веса точек (используется при построении регрессии или других статистических моделей).
* `row`: имя переменной, которая определяет, как данные будут разбиты на строки при использовании нескольких подграфиков.
* `col`: имя переменной, которая определяет, как данные будут разбиты на столбцы при использовании нескольких подграфиков.
* `col_wrap`: количество столбцов в сетке подграфиков до переноса на новую строку.
* `row_order`: порядок, в котором строки должны быть отображены.
* `col_order`: порядок, в котором столбцы должны быть отображены.
* `legend`: управляет отображением легенды (`'auto'`, `'brief'`, `'full'`, или `False`).
* `kind`: тип графика. Может быть `'scatter'` (по умолчанию) для диаграммы рассеяния или `'line'` для линейного графика.
* `height`: управляет высотой каждого подграфика (по умолчанию `height=5`).
* `aspect`: соотношение ширины и высоты каждого подграфика (по умолчанию `aspect=1`).

Для линейных графиков можно задействовать такие же, параметры как у функции `sns.lineplot()` (`marker`, `errorbar` и т.п.).

При использовании подграфиков можно перебирать их в цикле при помощи итератора `graph.axes.flat` (при условии сохранения графика-объекта в переменную `graph`). К примеру, для того, чтобы поворачивать метки по оси `x`, если они накладываются друг на друга (через комбинацию методов `get_xticklabels()` и `set_rotation()`, см. последний пример).

```python
# построение дефолтной диаграммы рассеяния через sns.relplot()
import matplotlib.pyplot as plt
import seaborn as sns

taxis = sns.load_dataset('taxis').dropna()
df = taxis[taxis['pickup_borough'] == 'Queens']

sns.relplot(data=df, x='pickup', y='total')

plt.show()
```

```python
# пример построения графика через sns.relplot() с контролем отношения высоты графика к его ширине
import matplotlib.pyplot as plt
import seaborn as sns

taxis = sns.load_dataset('taxis').dropna()
df = taxis[taxis['pickup_borough'] == 'Queens']

sns.relplot(data=df, x='pickup', y='total',
            aspect=2.2,
            hue='color',
            palette=['yellow', 'green'])

plt.show()
```

```python
# пример с сортировкой по дням недели
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

taxis = sns.load_dataset('taxis').dropna()
df = taxis[taxis['pickup_borough'] == 'Queens'].copy()
df['pickup_day_name'] = df['pickup'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['pickup_day_name'] = pd.Categorical(df['pickup_day_name'], categories=day_order, ordered=True)

sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'])

plt.show()
```

```python
# пример построения линейного графика через sns.relplot()
import matplotlib.pyplot as plt
import seaborn as sns

taxis = sns.load_dataset('taxis').dropna()
df = taxis[taxis['pickup_borough'] == 'Queens'].copy()
df['pickup_day_name'] = df['pickup'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['pickup_day_name'] = pd.Categorical(df['pickup_day_name'], categories=day_order, ordered=True)

sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'],
            kind='line')

plt.show()
```

```python
# пример изображения стиля линий в зависимости от категориальной переменной
import matplotlib.pyplot as plt
import seaborn as sns

taxis = sns.load_dataset('taxis').dropna()
df = taxis[taxis['pickup_borough'] == 'Queens'].copy()
df['pickup_day_name'] = df['pickup'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['pickup_day_name'] = pd.Categorical(df['pickup_day_name'], categories=day_order, ordered=True)

sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'],
            kind='line',
            style='color',
            dashes=[(2, 2), (6, 2)])

plt.show()
```

```python
# пример построения подграфиков на основе категориальной переменной
import matplotlib.pyplot as plt
import seaborn as sns

taxis = sns.load_dataset('taxis').dropna()

g = sns.relplot(data=taxis, x='pickup', y='total',
            height=4,
            hue='color',
            palette=['yellow', 'green'],
            col='pickup_borough',
            col_wrap=2)

for ax in g.axes.flat: # перебираем все подграфики
    for label in ax.get_xticklabels(): # перебираем метки на оси x
        label.set_rotation(45) # поворачиваем метки на оси x

plt.show()
```
